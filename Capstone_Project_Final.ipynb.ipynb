```python
# Task 2: Compute and replace missing values in 'Cost' column using running average with a For loop.

# --- Step 1: Load the Project DataFrame from the CSV file ---
# We are loading the project_data.csv file that was saved in Task 1.
# This ensures we use the saved file as per assignment instructions for Task 2 onwards.
project_df = pd.read_csv('project_data.csv')

# Display the DataFrame before imputation to confirm the missing values (NaN)
print("--- Project DataFrame before filling missing 'Cost' values ---")
print(project_df)
print("\n")
```

```python
# --- Step 2: Fill missing 'Cost' values using running average with a For loop ---

# Initialize a list to store non-missing cost values encountered so far
# This list will be used to calculate the 'running average'
non_missing_costs = []

# Iterate through each row of the 'project_df' using the index
# project_df.index gives us a list of numbers (0, 1, 2, ...) corresponding to each row
for i in project_df.index:
    # Get the current cost value from the 'Cost' column at the current row 'i'
    current_cost = project_df.loc[i, 'Cost']

    # Check if the current_cost is a missing value (NaN)
    # np.isnan() is used to check for NaN values because standard `== NaN` doesn't work
    if pd.isna(current_cost): # pd.isna() is a robust way to check for NaN in pandas
        # If it's a NaN, calculate the running average
        if non_missing_costs: # Check if there are any non-missing values recorded so far
            # Calculate the average of all costs collected in 'non_missing_costs'
            running_avg = sum(non_missing_costs) / len(non_missing_costs)
            # Replace the NaN value in the DataFrame with the calculated running average
            # project_df.loc[i, 'Cost'] allows us to select a specific cell by row index 'i' and column name 'Cost'
            project_df.loc[i, 'Cost'] = running_avg
            # Print a message to show what was filled, for verification
            print(f"Filled missing 'Cost' at index {i} with running average: {running_avg:.2f}")
        else:
            # If there are no non-missing costs yet (i.e., first value is NaN), we can't calculate an average.
            # For this assignment, we'll assume there's always an initial non-NaN value before a NaN.
            # If not, a common approach would be to fill with 0 or a global average later.
            project_df.loc[i, 'Cost'] = 0 # As a fallback, fill with 0 or a more suitable default if no previous values exist
            print(f"Filled missing 'Cost' at index {i} with 0 (no previous non-missing costs).")
    else:
        # If the current_cost is NOT a missing value, add it to our list of non-missing costs
        non_missing_costs.append(current_cost)

# Display the DataFrame after filling missing values to see the changes
print("\n--- Project DataFrame after filling missing 'Cost' values ---")
print(project_df)
print("\n")
```
